# **Revenue Optimization**

### Importing Libraries

```python
import pandas as pd
```

### Importing Dataset

```python
merged_df = pd.read_csv("../data/merged/merged_dataset.csv")
display(merged_df.head())
```

### Simulate Price Change

```python
optimized_df = merged_df.copy()

price_increase_pct = 5
optimized_df['optimized_price'] = ( optimized_df['price'] * (1 + price_increase_pct / 100) )
optimized_df['expected_units_sold'] = ( optimized_df['units_sold'] * 0.95 )
optimized_df['optimized_revenue'] = ( optimized_df['optimized_price'] * optimized_df['expected_units_sold'] )
```

### Result

```python
print( optimized_df[ [ 'product_name', 'price', 'optimized_price', 'optimized_revenue' ] ].head() )
```
