# **Customer Segmentation**

### Importing Libraries

```python
from sklearn.cluster import KMeans
import pandas as pd
```

### Importing Dataset

```python
merged_df = pd.read_csv("../data/merged/merged_dataset.csv")
display(merged_df.head())
```

### Customer Features

```python
customer_features = ( merged_df.groupby('segment') [['revenue', 'profit']] .sum() )
```

### KMeans Clustering

```python
model = KMeans( n_clusters=3, random_state=42, n_init=10 )
customer_features['cluster'] = ( model.fit_predict(customer_features) )
customer_features.reset_index(inplace=True)
```

### Result

```python
print(customer_features)
```
