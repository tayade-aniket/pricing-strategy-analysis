# **Pricing Elasticity Analysis**

### Importing Libraries

```python
import pandas as pd 
import numpy as np
```

### Importing Dataset

```python
merged_df = pd.read_csv("../data/merged/merged_dataset.csv")
display(merged_df.head())
```

### Price Elasticity Analysis

```python
elasticity_df = merged_df.copy()
```

### Sort Data Before Analysis

```python
elasticity_df = elasticity_df.sort_values(
    by='sales_date'
)
```

### Calculate Percentage Changes

```python
elasticity_df['price_change_pct'] = (
    elasticity_df['price']
    .pct_change()
)

elasticity_df['quantity_change_pct'] = (
    elasticity_df['units_sold']
    .pct_change()
)
```

### Calculate Elasticity

```python
elasticity_df['elasticity'] = (
    elasticity_df['quantity_change_pct'] /
    elasticity_df['price_change_pct']
)
```

### Handle Invalid Values

```python
elasticity_df.replace(
    [np.inf, -np.inf],
    0,
    inplace=True
)

elasticity_df.fillna(0, inplace=True)
```

### Elasticity Category

```python
elasticity_df['elasticity_category'] = (
    elasticity_df['elasticity']
    .apply(
        lambda x:
        'Highly Elastic'
        if abs(x) > 1
        else 'Inelastic'
    )
)
```

### Result

```python
display(
    elasticity_df[
        [
            'product_name',
            'price',
            'units_sold',
            'elasticity',
            'elasticity_category'
        ]
    ].head()
)
```
