# **Advanced Business Insights**

### Importing Libraries

```python
import pandas as pd
```

### Importing Dataset

```python
merged_df = pd.read_csv("../data/merged/merged_dataset.csv")
display(merged_df.head())
```

### Top Revenue Products

```python
revenue_insights = ( merged_df.groupby('product_name') ['revenue'] .sum() .sort_values(ascending=False) .head(10) )
print(revenue_insights)
```

### Most Profitable Category

```python
profit_insights = ( merged_df.groupby('category') ['profit'] .sum() .sort_values(ascending=False) )
print(profit_insights)
```

### Regional Perfomance

```python
regional_performance = ( merged_df.groupby('region') ['revenue'] .sum() .sort_values(ascending=False) )
print(regional_performance)
```

### Discount Impact

```python
merged_df['discount_bucket'] = pd.cut( merged_df['discount_percentage'], bins=[0, 5, 10, 20, 50], labels=[ 'Low', 'Medium','High', 'Extreme'])
impact_analysis = ( merged_df.groupby('discount_bucket') [['revenue', 'profit']] .mean() )
print(impact_analysis)
```
