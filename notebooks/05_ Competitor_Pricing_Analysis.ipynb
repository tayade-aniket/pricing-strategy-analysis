# **Competitor Pricing Analysis**

### Importing Libraries

```python
import pandas as pd
```

### Import Dataset

```python
merged_df = pd.read_csv("../data/merged/merged_dataset.csv")
display(merged_df.head())
```

### Compititor Analysis

```python
competitor_df = merged_df.copy()

competitor_df['price_gap'] = (competitor_df['price']-competitor_df['competitor_price'] )
competitor_df['pricing_position'] = ( competitor_df['price_gap'] .apply( lambda x: 'Premium' if x > 0 else 'Competitive' ) )
```

### Result

```python
print( competitor_df[ [ 'product_name', 'price', 'competitor_price', 'price_gap', 'pricing_position' ] ].head() )
```
