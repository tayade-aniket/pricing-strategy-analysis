# **Data Loading and Preprocessing**

### Importing Libraries

```python
import pandas as pd
import numpy as np
```

### Loading Datasets

```python
retail_df = pd.read_csv("../data/raw/retail_price_optimization.csv")
display(retail_df)
```

```python
superstore_df = pd.read_csv(
    "../data/raw/superstore.csv",
    encoding="latin1"
)

display(superstore_df)
```

```python
survey_df = pd.read_csv("../data/raw/retail_price_survey.csv")
display(survey_df)
```

### Standardize Columns

```python
def standardize_columns(df):

    df.columns = (
        df.columns
        .str.lower()
        .str.strip()
        .str.replace(' ', '_')
        .str.replace('-', '_')
    )

    return df


retail_df = standardize_columns(retail_df)
superstore_df = standardize_columns(superstore_df)
survey_df = standardize_columns(survey_df)
```

### Handle Missing Values

```python
retail_df.ffill(inplace=True)
superstore_df.ffill(inplace=True)
survey_df.ffill(inplace=True)
```

### Convert Dates

```python
superstore_df['order_date'] = pd.to_datetime(
    superstore_df['order_date'],
    dayfirst=True
)

superstore_df['ship_date'] = pd.to_datetime(
    superstore_df['ship_date'],
    dayfirst=True
)

survey_df['date'] = pd.to_datetime(
    survey_df['date'],
    dayfirst=True
)
```
